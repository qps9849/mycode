```python
train_cols=df.columns[0:8]
X=df[train_cols]
y=df["outcome"]

print(type(X), type(y))
print(X.head())
print(y.head())
```

```python
import mglearn
pd.plotting.scatter_matrix(df,c=y,figsize=(15,15),marker='o', cmap=mglearn.cm3)
```

```python
from imblearn.under_sampling import RandomUnderSampler

X_sample,y_sample= RandomUnderSampler(random_state=0).fit_resample(X,y)
X_samp= pd.DataFrame(data=X_sample, columns=train_cols)
y_samp=pd.DataFrame(data=y_sample, columns=['outcome'])
df_samp=pd.concat([X_samp,y_samp],axis=1)
```

```python
from imblearn.over_sampling import SMOTE

X_sample,y_sample= SMOTE(random_state=0).fit_resample(X,y)
X_samp= pd.DataFrame(data=X_sample, columns=train_cols)
y_samp=pd.DataFrame(data=y_sample, columns=['outcome'])
df_samp=pd.concat([X_samp,y_samp],axis=1)
```

```python
df_samp['outcome'].value_counts()
```

```python
X=df_samp[train_cols]
y=df_samp["outcome"]
```

```python
from sklearn.model_selection import  train_test_split

X_train,X_test, y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=10)
```

```python
from tensorflow.keras.models import Sequential
from keras.layers.core import Dense
model=Sequential()
model.add(Dense(128,input_shape=(len(X_train.columns),),activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='mse', optimizer='adam',metrics=['accuracy'])
```

```python
import pandas as pd
pd.DataFrame(X_train).boxplot()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
pd.DataFrame(X_train_scaled).boxplot()
X_train_scaled=pd.DataFrame(X_train_scaled)
X_test_scaled=pd.DataFrame(X_test_scaled)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping=EarlyStopping(patience=2)
hist=model.fit(X_train_scaled, y_train, epochs=500,
validation_split=0.2, callbacks=[early_stopping])
```

```python
plt.figure(figsize=(12,8))
plt.plot(hist.history['loss'])
plt.plot(hist.history['accuracy'])
plt.legend(['loss','accuracy'])
plt.grid()
plt.show()
```

```python
scores = model.evaluate(X_train_scaled, y_train, verbose=1)
print(scores)
print("%s: %.2f%%" %(model.metrics_names[1],scores[1]*100))

scores = model.evaluate(X_test_scaled, y_test, verbose=1)
print(scores)
print("%s: %.2f%%" %(model.metrics_names[1],scores[1]*100))

pred=model.predict(X_test_scaled)

import numpy as np

pred=model.predict(X_test_scaled)
np.argmax(pred[:5],axis=1)
```
