```python
#숫자의 자리수
def union(a,b):
    c=[]
    for i in a:
        if i not in b:
            c.append(i)
    c=c+b
    c.sort()
    return c
a=[1,3,4,5,6]
b=[2,1,6,3,9]
union(a,b)
```

```python
def intersection(a,b):
    c=[]
    for i in a:
        if i in b:
            c.append(i)
    c.sort()
    return c
a=[1,3,4,5,6]
b=[2,1,6,3,9]
intersection(a,b)
```

```python
def complement(a,b):
    c=a+[]
    for i in b:
        if i in a:
            c.remove(i)
    c.sort()
    return c
a=[1,3,4,5,6]
b=[2,1,6,3,9]
complement(a,b)
```

```python
def rotate(items):
    result=[]
    result.append(items[-1])
    result=result+(items[:-1])
    return result
a=[1,2,3,4]
rotate(a)
```

```python
def my_zip(a,b):
    if len(a)!=len(b):
        print('두 입력값의 길이가 다릅니다')
    else:
        n=len(a)
        c=[]
        for i in range(n):
            c.append((a[i],b[i]))
        return c
a=[1,2,3]
b=['a','b','c']
my_zip(a,b)
```

```python
def saving(money,years,rate):
    result = money*(1+rate/100)**years
    return result
```

```python
money=10000
years=np.arange(0,11)
rate=[3,5,7]
total_money=np.zeros((len(rate),len(years)))
for i in range(len(rate)):
    for j in range(len(years)):
        total_money[i,j]=saving(money,years[j],rate[i])
    plt.plot(years, total_money[i,:],label='{}%'.format(rate[i]))
    
plt.legend(fontsize=14)
plt.grid()
plt.xlabel('Years',fontsize=14)
plt.ylabel('Total money', fontsize=14)
```
