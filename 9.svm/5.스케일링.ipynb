```python
import pandas as pd
df=pd.read_csv('c:/data/rides/rides2.csv')
df.columns
```

```python
cols=['weekend', 'num.child', 'distance', 'rides', 'games', 'wait', 'clean']
X=df[cols]
X
```

```python
X.boxplot()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(X)
df_ss=pd.DataFrame(ss.transform(X))
df_ss
```

```python
df_ss.boxplot()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
mms.fit(X)
df_mms=pd.DataFrame(mms.transform(X))
df_mms
```

```python
df_mms.boxplot()
```

```python
print(mms.data_max_)
print(mms.data_min_)

X2=mms.transform(X)
print(X2)
print(mms.inverse_transform(X2))
```
