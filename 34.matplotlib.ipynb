```python
from sklearn import datasets
iris=datasets.load_iris()
iris.target
```

```python
import numpy as np
plt.figure(figsize=(14,7))
colormap=np.array(['red','lime','black'])
plt.subplot(1,2,1)
plt.scatter(X['sepal length (cm)'],X['sepal width (cm)'],c=colormap[iris.target],s=40)
plt.title('꽃받침')
plt.subplot(1,2,2)
plt.scatter(X['petal length (cm)'],X['petal width (cm)'],c=colormap[iris.target],s=40)
plt.title('꽃잎')
```
